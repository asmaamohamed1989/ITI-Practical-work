#### Use your code from practical session 1 and create a function to perform GD for single variable linear regression.
#### The function should take the following input parameters:
##### Input data (X), Target Labels (y), Learning Rate (alpha), Maximum number of iterations
#### The function should return the following:
##### All Theta_0 and Theta_1 values till converge, and equivellant Losses, all hypothesis output, and optimum values of Theta_0 and Theta_1
#### Note: you can choose any suitable stop criteria.

```python
def Gradient_Descent(x_i,y_i,alpha,iterations):
    theta_0= 0
    theta_1= 0
    alpha =alpha
   
    cost=[]
    All_theta_0 =[]
    All_theta_1 =[]
    All_hypothesis =[]
    
    max_iterations =1700
    
    for i in range(0,max_iterations):
        
        hypothesis = theta_0 + theta_1 * x_i
        All_hypothesis.append(hypothesis)
        
        All_theta_0.append(theta_0)
        All_theta_1.append(theta_1)
        
        cost_function = (sum(hypothesis - y_i)**2) / (2*len(x_i))
        cost.append(cost_function)
        
        
        derivative_y =  np.sum(hypothesis - y_i) / (len(x_i))
        derivative_x =  np.sum((hypothesis - y_i)* x_i) / (len(x_i))
        
        theta_0 = theta_0 - alpha * derivative_y
        theta_1 = theta_1 - alpha * derivative_x
        
        optimum_theta0 = theta_0
        optimum_theta1 = theta_1
        
        
        if (cost_function < 0.0001):
            break
          
        
        
    return All_theta_0,All_theta_1,cost,All_hypothesis,optimum_theta0,optimum_theta1,hypothesis
        
        
        
        
        
        
```

#### Update the implementation of the Batch GD to be Mini-Batch and rpeat the above plots and evaluations
#### Change batch size and see the effect

```python
def Mini_Batch_Gradient_Descent (x_i, y_i, learning_rate, max_iterations, batch_size):
#batch_size =10
    theta_0 = 0     # assumed value of the parameters
    theta_1 = 0
    learning_rate = learning_rate
    cost = []
    All_theta_0 = []
    All_theta_1 =[]
    All_Hypothesis =[]
    iter =0
    x_i_batch=[]
    y_i_batch=[]
    Hypothesis_list = []
    
    for k in range(max_iterations):
        # range(0,50,10) the number of batches =50/10=5
        for i in range (0,len(x_i),batch_size):
             
            if i == len(x_i):
                Break
            for j in range(i,i+batch_size):
            
                x_i_batch.append(x_i[j])
                y_i_batch.append(y_i[j])
            
            
           
            x_i_batch1 = np.array(x_i_batch)
            y_i_batch1 = np.array(y_i_batch)
        
            
        
            hypothesis = theta_0 + theta_1* x_i_batch1
            All_Hypothesis.append(hypothesis)
        
            All_theta_0.append(theta_0)
            All_theta_1.append(theta_1)
        
            cost_function = (sum (hypothesis - y_i_batch1)**2 )/ (2*(batch_size))
            cost.append(cost_function)
    
            theta0_gradient = np.sum (hypothesis - y_i_batch1) / batch_size
            theta1_gradient = np.sum ((hypothesis - y_i_batch1)*x_i_batch1) / batch_size
    
            theta_0 = theta_0 - learning_rate * theta0_gradient
            theta_1 = theta_1 - learning_rate * theta1_gradient
            
            x_i_batch.clear()
            y_i_batch.clear()
        
        iter+=1
        if (cost_function < 0.01):
            break
       
            
        Hypothesis = theta_0 + theta_1*x_i
        Hypothesis_list.append(Hypothesis)
    optimum_theta0 = theta_0
    optimum_theta1 = theta_1
    hypothesis = optimum_theta0 + optimum_theta1* x_i
    
    return All_theta_0,All_theta_1, cost , All_Hypothesis, optimum_theta0, optimum_theta1, hypothesis, iter,  Hypothesis_list
```
