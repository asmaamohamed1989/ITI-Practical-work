```python
#The point geom is used to create scatterplots. The scatterplot is most useful for displaying the relationship between two continuous variables.
ggplot(df, aes(x = "alcohol", y = "proline")) + geom_point()
```

```python
pca = PCA()
pca.fit(df[features])
#explained_variance_ratio_ method of PCA is used to get the ration of variance (eigenvalue / total eigenvalues) 
pca.explained_variance_ratio_
```

```python
pcaDF = pd.DataFrame({"pc": range(1, 14),"expl_var" : pca.explained_variance_ratio_,  "cum_var": pca.explained_variance_ratio_.cumsum()})
pcaDF.head()
```

```python
(ggplot(pcaDF, aes(x = "pc", y = "expl_var")) + geom_line() + geom_point())
```

```python
(ggplot(pcaDF, aes(x = "pc", y = "cum_var")) + geom_line(color = "pink") +
 geom_point(color = "pink") + geom_hline(yintercept = 0.95))
```
