```python
# K
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=4)
kmeans.fit(dataset1[["x","y"]])
y_kmeans = kmeans.predict(dataset1[["x","y"]])
print(y_kmeans)

##Calculate Silhouette Score for K-Means Clusters 

score = silhouette_score(dataset1[["x","y"]], y_kmeans, metric='euclidean')
#
# Print the score
#
print('Silhouetter Score: %.3f' % score)
```

```python
plt.scatter(dataset1["x"], dataset1["y"], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
# Compute DBSCAN
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.3, min_samples=10).fit(dataset1[["x","y"]])
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
import sklearn.metrics as metrics
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print('Estimated number of clusters: %d' % n_clusters_)
print('Estimated number of noise points: %d' % n_noise_)

print("Silhouette Coefficient: %0.3f" 
      % metrics.silhouette_score(dataset1[["x","y"]], db.labels_))
```
