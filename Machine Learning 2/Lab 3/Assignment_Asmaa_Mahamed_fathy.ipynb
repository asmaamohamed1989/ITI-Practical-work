```python
#!pip install pandas plotnine
# use any plotting library seaborn matplotlib,..........
from sklearn.preprocessing import StandardScaler
import pandas as pd
from plotnine import *
from sklearn.neighbors import NearestNeighbors
import sklearn
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import matplotlib.cm
```

```python
samples = 800

data1 = sklearn.datasets.make_blobs(n_samples=samples, random_state=42)
data1=pd.DataFrame(data1[0])

data2= pd.read_csv("dataset5.csv",header=None)
```

```python
list_of_datasets=[data1,data2]

for data in list_of_datasets:
    standard = StandardScaler()
    data[["x","y"]] = standard.fit_transform(data)
    print(ggplot(data, aes(x = "x", y = "y")) + geom_point())
```

# K-means

```python
from sklearn.cluster import KMeans

for data in list_of_datasets:
  km=KMeans(n_clusters=5, random_state=42)
  clusters=km.fit_predict(data[['x','y']])
  centers=km.cluster_centers_
  unique_clusters=np.unique(clusters) 
#   print(km.inertia_)
  for uc in unique_clusters:
    plt.scatter(data['x'],data['y'],c=clusters)
  plt.scatter(centers[:,0],centers[:,1],s=73,marker='x',c='r')
  plt.show()
```

# Hierarchical

```python
import scipy.cluster.hierarchy as shc
for i in list_of_datasets:
  plt.figure(figsize=(20,8))
  dendrogram = shc.dendrogram(shc.linkage(y=i[['x','y']] , method='average'))
```

## Agglomerative

```python
from sklearn.cluster import AgglomerativeClustering
for i in list_of_datasets:
  agg_cluster=AgglomerativeClustering(linkage='single')
  labels=agg_cluster.fit_predict(i[['x','y']])
  plt.scatter(i['x'],i['y'],c=labels)
  plt.show()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
# from sklearn.metrics import silhouette_score
db=DBSCAN(eps=0.2, min_samples=5)
for i in list_of_datasets:
  labels=db.fit_predict(i[['x','y']])
  plt.scatter(i['x'],i['y'],c=labels)
  plt.show()
```

# GMM-EM

```python
from sklearn.mixture import GaussianMixture

for i in list_of_datasets:
    gmm = GaussianMixture(n_components=5)
    labels=gmm.fit_predict(i)
    plt.scatter(i['x'],i['y'],c=labels)
    plt.show()
```
