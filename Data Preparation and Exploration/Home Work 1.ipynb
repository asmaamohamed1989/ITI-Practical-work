# Explore the data and prepare it for visualization


```python
import matplotlib.pyplot as plt
# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(df['dateRep'], df['deaths'])
# set a title and labels
ax.set_title('COVID-19 pandemic dataset')
ax.set_xlabel('dateRep')
ax.set_ylabel('deaths')
```

```python
df.groupby("countriesAndTerritories").cases.mean().sort_values(ascending=False)[:5].plot.bar()
```

```python
import seaborn as sns
sns.heatmap(df.corr(), annot=True)
```

```python
sns.pairplot(df)
```
