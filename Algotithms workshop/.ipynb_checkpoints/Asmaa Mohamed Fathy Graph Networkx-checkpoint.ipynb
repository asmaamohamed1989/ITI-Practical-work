```python
import networkx as nx
G = {   "0": ["1", "3" , "8"],
            "1": ["0","7"],
            "2": ["3","7","5"],
            "3": ["0","4"],
            "4": ["3","8"],
            "5": ["2","6"],
            "6": ["5"],
            "7": ["1","2"] }

graph = nx.Graph(G) #create a graph

```

```python
import matplotlib.pyplot as plt

nx.draw(graph, with_labels=True, font_weight='bold')

plt.show()
```

```python
dfs_output = list(nx.dfs_preorder_nodes(graph, source="0"))

print(dfs_output)
```

```python
Bfs_output = list(nx.bfs_tree(graph, source="0"))

print(Bfs_output)
```

```python
print(nx.dijkstra_path(graph, "0", "4"))
```

```python
graph = {
    0: [2,1],
    1: [0,6,5,3],
    2: [5,6,0],
    3: [1,7],
    4: [6,7],
    5: [2,7,1],
    6: [2,1,7,4],
    7: [4,6,5,3],
    }
G = nx.Graph(graph)

G.add_edge(0,1,weight=8)
G.add_edge(0,2,weight=9)
G.add_edge(1,3,weight=1)
G.add_edge(1,5,weight=7)
G.add_edge(1,6,weight=9)
G.add_edge(2,5,weight=6)
G.add_edge(2,6,weight=4)
G.add_edge(3,7,weight=4)
G.add_edge(4,7,weight=7)
G.add_edge(5,7,weight=2)
G.add_edge(6,7,weight=6)


pos = nx.spring_layout(G)  # positions for all nodes

# nodes
nx.draw_networkx_nodes(G, pos, node_size=700)

labels = nx.get_edge_attributes(G,'weight')

# edges
nx.draw_networkx_edges(G, pos, width=2)
nx.draw_networkx_edges( G, pos, width=6, alpha=0.5, edge_color="b",style="dashed")

# labels
nx.draw_networkx_labels(G, pos, font_size=10)

nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)


plt.axis("off")
plt.show()
```

```python
distance, path = nx.single_source_dijkstra(G, source=0, target=6)
print(distance, path)
```
