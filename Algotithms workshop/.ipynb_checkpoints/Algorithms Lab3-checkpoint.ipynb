```python
class Stack:
    
    
    def __init__(self, the_stack_length):
        self.the_stack_length = the_stack_length
        self.arr = []
    def push(self,new_value):
        if self.is_full():
            print('stack is full')
        else:
            self.arr.append(new_value)
        
    def pop(self):
        if self.is_empty():
            print('there is nothing to pop, stack is EMPTY!!')
        else:
            last_element = self.arr[len(self.arr)-1]
            del(self.arr[len(self.arr)-1])
            return last_element
        
    def is_full(self):
        if len(self.arr) == self.the_stack_length:
            return True
        else:
            return False
            
    def is_empty(self):
        if len(self.arr) == 0:
            return True
        else:
            return False

s = Stack(7)
s.arr = [2,4,6,8]
print(s.pop())
print(s.pop())
print(s.pop())
print(s.pop())

print(s.pop())

print(s.arr)
```

```python
class Queue:
    
    
    def __init__(self, the_stack_length):
        self.the_stack_length = the_stack_length
        self.arr = []
    def push(self,new_value):
        if self.is_full():
            print('Queue is full')
        else:
            self.arr.append(new_value)
        
    def pop(self):
        if self.is_empty():
            print('there is nothing to pop, Queue is EMPTY!!')
        else:
            first_element = self.arr[0]
            del(self.arr[0])
            return first_element
        
    def is_full(self):
        if len(self.arr) == self.the_stack_length:
            return True
        else:
            return False
            
    def is_empty(self):
        if len(self.arr) == 0:
            return True
        else:
            return False

q = Queue(7)
q.arr = [2,4,6,8]
print(q.pop())
print(q.pop())
print(q.pop())
print(q.pop())

print(q.pop())

print(q.arr)
```

```python
graph = {
    'a': ['b','c','e'],
    'b': ['a','d'],
    'c': ['a','e','d','f'],
    'd': ['b','c','e'],
    'e': ['a','c','d'],
    'f': ['c'],
    }
# print(graph['a'])

def bfs(graph,start):
    list_of_nodes = list(graph.keys())
    nodes_to_visit_queue = Queue(len(list_of_nodes))
    nodes_to_visit_queue.push(start)
    visited_nodes = []
    inner_list = []
    for i in range(len(list_of_nodes)):
        if nodes_to_visit_queue.is_empty():
            break
        start = nodes_to_visit_queue.pop()
        print('start',start)
        inner_list = list(graph[start])
        visited_nodes.append(start)
        
        print('visited_nodes',visited_nodes)
        
        for j in inner_list:
            if j not in visited_nodes and j not in nodes_to_visit_queue.arr:
                nodes_to_visit_queue.push(j)
        print('nodes_to_visit_stack',nodes_to_visit_queue.arr)
                
bfs(graph,'a')        
    
```
