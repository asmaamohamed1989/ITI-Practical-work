```python
from pysm import State, StateMachine, Event

def getNewState(currentStateName, option):
    fs, boat, ss = currentStateName.split("|")
    boatFS = int(boat[0])
    boatSS = int(boat[1])
    first_Side_Guards_Numbers = int(fs[0])
    first_Side_prisoners_Numbers = int(fs[3])
    second_Side_Guards_Numbers = int(ss[0])
    second_Side_prisoners_Numbers = int(ss[3])
        
    if option == 'G':
        if boatFS == 1:
            if first_Side_Guards_Numbers >= 1:
                first_Side_Guards_Numbers-=1
                second_Side_Guards_Numbers+=1
                boatFS = 0
                boatSS = 1   
        else:
            if second_Side_Guards_Numbers >= 1:
                second_Side_Guards_Numbers-=1
                first_Side_Guards_Numbers+=1
                boatFS = 1
                boatSS = 0
                
    elif option == 'GG':
        if boatFS == 1:
            if first_Side_Guards_Numbers >= 2:
                first_Side_Guards_Numbers-=2
                second_Side_Guards_Numbers+=2
                boatFS = 0
                boatSS = 1
        else:
            if second_Side_Guards_Numbers >= 2:
                second_Side_Guards_Numbers-=2
                first_Side_Guards_Numbers+=2
                boatFS = 1
                boatSS = 0
                
    elif option == 'GP':
        if boatFS == 1:
            if first_Side_Guards_Numbers >= 1:
                first_Side_Guards_Numbers-=1
                first_Side_prisoners_Numbers-=1     
                second_Side_Guards_Numbers+=1 
                second_Side_prisoners_Numbers+=1  
                boatFS = 0
                boatSS = 1
        else:
            if second_Side_Guards_Numbers >= 1:
                second_Side_Guards_Numbers-=1 
                second_Side_prisoners_Numbers-=1     
                first_Side_Guards_Numbers+=1
                first_Side_prisoners_Numbers+=1
                boatFS = 1
                boatSS = 0
        
    elif option == 'PP':
        if boatFS == 1:
            if first_Side_prisoners_Numbers >= 2:
                first_Side_prisoners_Numbers-=2
                second_Side_prisoners_Numbers+=2
                boatFS = 0
                boatSS = 1                    
        else:
            if second_Side_prisoners_Numbers >= 2:
                second_Side_prisoners_Numbers-=2
                first_Side_prisoners_Numbers+=2
                boatFS = 1
                boatSS = 0

    elif option == 'P':
        if boatFS == 1:
            if first_Side_prisoners_Numbers >= 1:
                first_Side_prisoners_Numbers-=1
                second_Side_prisoners_Numbers+=1
                boatFS = 0
                boatSS = 1
        else:
            if second_Side_prisoners_Numbers >= 1:
                second_Side_prisoners_Numbers-=1
                first_Side_prisoners_Numbers+=1
                boatFS = 1
                boatSS = 0
                
    return ('{}G,{}P|{}{}|{}G,{}P'.format(first_Side_Guards_Numbers, first_Side_prisoners_Numbers, boatFS, boatSS, second_Side_Guards_Numbers, second_Side_prisoners_Numbers))

def isValidState (stateName):
    fs, boat, ss = stateName.split("|")
    boatFS = int(boat[0])
    boatSS = int(boat[1])
    first_Side_Guards_Numbers = int(fs[0])
    first_Side_prisoners_Numbers = int(fs[3])
    second_Side_Guards_Numbers = int(ss[0])
    second_Side_prisoners_Numbers = int(ss[3])    
    
    if (0 < first_Side_Guards_Numbers < first_Side_prisoners_Numbers) | (0 < second_Side_Guards_Numbers < second_Side_prisoners_Numbers):
        return False
    return True

def createFSM(states, stateMachine):
    for i in range(len(states)):
        states.insert(i, State(states.pop(i)))
        stateMachine.add_state(states[i])
    stateMachine.initialize()
def createTransations(states, stateMachine):
    print("============================== Transition =================================")
    for i in range(0,12):
        stateMachine.add_transition(states[i%12], states[(i+1)%12], events=[states[(i+1)%12].name])
        print("1st State -> {} 2nd State -> {} event -> {}".format(states[i%12], states[(i+1)%12], states[(i+1)%12].name))

def generateStates(States):
    popedStates = []
    options = ['P','PP','GP','GG','G']
    states_number = 0
    fs, boat, ss = States[states_number].split("|")
    while(int(ss[0]) + int(ss[3]) < 6):
        beforeGettingNewState = states_number
        for option in options:
            returnedState = getNewState(states[states_number], option)
            if isValidState(returnedState):
                if(returnedState not in States) & (returnedState not in popedStates):
                    States.append(returnedState)
                    states_number+=1
                    break
        if beforeGettingNewState == states_number:
            popedStates.append(States.pop())
            states_number-=1 
        fs, boat, ss = States[states_number].split("|")

print(" ============================== Start Creating State Machine =================================")   
sm = StateMachine('sm')
first = State('3G,3P|10|0G,0P') #format that represent the states (G,P|Boat side|G,P) first pip is the first coast and the other one is the second
sm.add_state(first, initial=True)
states =[first.name]
print("...............................................................................................\n")
generateStates(states)
states.pop(0)  # poped becasue it's already in the state machine as intial state
createFSM(states, sm)

print("============================== State Machine States generated =================================\n")   

states.insert(0, first) # so i can access the state machine using the state name (as transition event)

createTransations(states, sm)

print("============================== StateMachine =================================")
for i in range(0, 12):
    print("State Machine Current State >>>> [{}]".format(sm.state))
    assert sm.state == states[i%12]
    sm.dispatch(Event(states[(i+1)%12].name))
```
